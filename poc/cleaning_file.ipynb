```python
import pandas as pd
import numpy as np
```

```python
# Load in data for further exploration.
df = pd.read_csv("../files/oh11.csv")
```

```python
# View the first 5 rows of the df.
df.head()
```

```python
# Get the basic info on df columns, dtypes, and coverage.
df.info()
```

```python
# Look at the basic df stats.
df.describe()
```

```python
# Confirm the data types found in each column.
df.applymap(type).apply(pd.value_counts).fillna(0)
```

```python
# Looks like the zip column has string values instead of numeric.
df[df.zip.apply(type) == str]
```

```python
# Found unusual zip values (i.e. "NO_ZIP!") - replace these for NaN values.
df["zip"].replace({"NO_ZIP!": ""}, inplace=True)

# Recast zip column to string not that the weird values are cleaned.
df["zip"] = df.zip.apply(lambda x: str(x).zfill(5))
```

```python
# Confirm that the zip range is valid for the state/district in the voter file.
print(min(df["zip"]))
print(max(df["zip"]))
df["zip"].str.count("00000").sum()
df["zip"].replace({"00000": np.nan}, inplace=True)
```

```python
# Count the number of NaN values in every column. Sanity check based on expectations from vf.
df.isna().sum()
```

```python
# Confirm the data types found in each column.
df.applymap(type).apply(pd.value_counts).fillna(0)
```

```python
df.head()
```

```python
# Count the number of duplicates in the df.
df_dups = df[df.duplicated()].reset_index()
df_dups.info()
```

```python
df_no_dups = df[~df.duplicated()].reset_index(drop=True)
df_no_dups.info()
```

```python
# Looks like there are empty values for state when there shouldn't be. Clean this!
df_no_dups[pd.isna(df_no_dups["state"])]
df_no_dups["state"] = df_no_dups["state"].fillna("OH")
```

```python
# Count the number of NaN values in every column. Sanity check based on expectations from vf.
df_no_dups.isna().sum()
```

```python
# Inspect the date columns and make sure they are reasonable min/max values.
df_no_dups["date_of_birth"] = pd.to_datetime(df["date_of_birth"], format='%Y-%m-%d')
df_no_dups["registration_date"] = pd.to_datetime(df["registration_date"], format='%Y-%m-%d')
print(min(df_no_dups["date_of_birth"]))
print(max(df_no_dups["date_of_birth"]))
print(min(df_no_dups["registration_date"]))
print(max(df_no_dups["registration_date"]))
```

```python
# Unexpectedly, there are (4) birthdates from 1800-01-01 which is impossible. 
df_no_dups[df_no_dups["date_of_birth"].astype(str).str.contains("1800")]
```

```python
# Assumption: this dob is likely missing and using a default value (reg_date looks reasonable), so updating dob.
df_no_dups = df_no_dups.replace("1800-01-01", "1900-01-01")
print(min(df_no_dups["date_of_birth"]))
```

```python
# Confirm the data types found in each column.
df_no_dups.applymap(type).apply(pd.value_counts).fillna(0)
```

```python
# Check name fields for punctuation or numeric chars.
punctuation = "!\"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~1234567890"
df_no_dups[df_no_dups["first_name"].map(lambda v: any(char in v for char in punctuation))]
```

```python
# Check name fields and confirm NaN values present.
df_no_dups[df_no_dups["last_name"].map(lambda v: isinstance(v, float))]
```

```python
# Observation: there are first_name values that are a single character (i.e. "A") - depending on how we are using the
# voter file we might not want to use these records (and scrub them out entirely), but these program specs can be 
# accommodated for by using filters. 

# Assumption: I am assuming that we will want to maintain these records as a base table and filter based on 
# program requirements (i.e. mail programs will want properly cleaned first_name records only).
df_no_dups["first_name"].replace({".": np.nan}, inplace=True)
```

```python
# Count the number of NaN values in every column. Sanity check based on expectations from vf.
df_no_dups.isna().sum()
```

```python
# Export cleaned df
df_no_dups.to_csv("../files/oh_cd11_clean.csv", index=False)
print(f"Data record count: {len(df_no_dups)}")
```
